```python
import pandas as pd

data = {
    'Sales Rep': ['bob ', 'DAVID', '  Charlie'],
    'Region': ['north', 'South ', ' EAST'],
    'Product Category': ['furniture', 'ELECTRONICS ', ' food'],
    'Customer Type': ['returning', ' NEW', 'Returning']
}

df = pd.DataFrame(data)
print(" Before Cleaning:\n", df)
```

```python


# Standardize the text values like country,names etc
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print("\n✅ Cleaned column names:")
print(df.columns.tolist())

# Step 4: Standardize text columns using cleaned names
try:
    df['sales_rep'] = df['sales_rep'].str.strip().str.title()
    df['region'] = df['region'].str.strip().str.title()
    df['product_category'] = df['product_category'].str.strip().str.title()
    df['customer_type'] = df['customer_type'].str.strip().str.capitalize()
    print("\n✅ Successfully cleaned text columns.")
except KeyError as e:
    print(f"\n❌ Column not found: {e}")
```

```python
# Convert 'sale_date' to datetime format to a consistent type
df['sale_date'] = pd.to_datetime(df['sale_date'], errors='coerce', dayfirst=True)
print("\n  Nulls after date conversion:", df['sale_date'].isnull().sum())
df = df.dropna(subset=['sale_date'])
print("\n Sample of cleaned dates:\n", df[['sale_date']].head())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# Rename column headers to be clean and uniform
print("\n Renamed Columns:\n", df.columns.tolist())
```

```python
# Check & fix datatypes
df['quantity_sold'] = pd.to_numeric(df['quantity_sold'], errors='coerce')
df['unit_cost'] = pd.to_numeric(df['unit_cost'], errors='coerce')
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
df['sales_amount'] = pd.to_numeric(df['sales_amount'], errors='coerce')
df = df.dropna(subset=['quantity_sold', 'unit_cost', 'unit_price', 'sales_amount'])
print("\n Data types after conversion:\n")
print(df.dtypes)
```

```python
# Save cleaned data to a new CSV file
df.to_csv("cleaned_sales_data.csv", index=False)

print("\n  Cleaned data saved successfully as 'cleaned_sales_data.csv'")
```
